#  **Prepare both the model and tokenizer**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "mistralai/Mistral-7B-Instruct-v0.2"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    load_in_4bit=True
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

#  **Extract text from a youtube video's audio**

```python
import subprocess

# download the audio file of a youtube video
subprocess.run([
    "yt-dlp",
    "-x", "--audio-format", "mp3",
    "-o", "audio.mp3",
    "https://www.youtube.com/watch?v=x7X9w_GIm1s".split('&')[0]
])
```

```python
import whisper

audio2text = whisper.load_model("base").to("cuda")
result = audio2text.transcribe("audio.mp3")

result['text']
```

#  **Prepare the input and use the model**

```python
prompt = f"""### Instruction:
Summarize the following lecture as a set of bullet points. For each bullet point, write one or more detailed paragraphs to explain it clearly, including any relevant concepts, examples, or context. Also write everything in a format that can be pasted directly on Notion:

{result['text']}

### Response:
"""

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=10000,
    do_sample=True,
    temperature=0.7
)

tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
prompt = f"""### Instruction:
Generate as many as needed quiz questions with answers about the following text. Also write everything in a format that can be pasted directly on Notion:

{result['text']}

### Response:
"""

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=2000,
    do_sample=True,
    temperature=0.7
)

tokenizer.decode(outputs[0], skip_special_tokens=True)
```

#  **Check if the video is about python. If so, write a notebook**

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

res = classifier("this video explains how to use pandas", ["python", "not python"])
res
```

```python
if res['labels'][0] == 'python':
  prompt = f"""### Instruction:
  Given this programming lecture, write a Jupyter Notebook with exercises and solutions:

  {result['text']}

  ### Response:
  """

  inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

  outputs = model.generate(
      **inputs,
      max_new_tokens=2000,
      do_sample=True,
      temperature=0.7
  )

  print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
